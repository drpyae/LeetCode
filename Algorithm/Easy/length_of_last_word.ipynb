# Length of Last Word
Given a string s consisting of words and spaces, return the length of the last word in the string.

A word is a maximal 
substring
 consisting of non-space characters only.

 

Example 1:

Input: s = "Hello World"
Output: 5
Explanation: The last word is "World" with length 5.
Example 2:

Input: s = "   fly me   to   the moon  "
Output: 4
Explanation: The last word is "moon" with length 4.
Example 3:

Input: s = "luffy is still joyboy"
Output: 6
Explanation: The last word is "joyboy" with length 6.
 

Constraints:

1 <= s.length <= 104
s consists of only English letters and spaces ' '.
There will be at least one word in s.

To solve this problem, we need to find the length of the last word in the given string `s`. A word is defined as a maximal substring consisting of non-space characters.

### Approach:

1. **Trim the String**:
   - Remove any leading or trailing spaces in the string `s` using the `strip()` method. This helps eliminate any unnecessary spaces that could affect our result.

2. **Split the String into Words**:
   - Use the `split()` method to split the trimmed string by spaces. This method returns a list of words (substrings separated by spaces).

3. **Find the Last Word**:
   - The last word in the list can be accessed using `words[-1]` (the last element of the list).

4. **Return the Length of the Last Word**:
   - Return the length of the last word using the `len()` function.

### Python Implementation:

```python
def lengthOfLastWord(s):
    # Trim the string to remove leading and trailing spaces
    s = s.strip()

    # Split the string into words
    words = s.split()

    # Return the length of the last word
    return len(words[-1])
```

### Example Usage:

```python
# Example 1
s1 = "Hello World"
print(lengthOfLastWord(s1))  # Output: 5

# Example 2
s2 = "   fly me   to   the moon  "
print(lengthOfLastWord(s2))  # Output: 4

# Example 3
s3 = "luffy is still joyboy"
print(lengthOfLastWord(s3))  # Output: 6
```

### Explanation of Examples:

1. **Example 1**:
   - Input: `s = "Hello World"`
   - The last word is `"World"` with a length of `5`.
   - Output: `5`

2. **Example 2**:
   - Input: `s = "   fly me   to   the moon  "`
   - After trimming, the string becomes `"fly me   to   the moon"`.
   - The last word is `"moon"` with a length of `4`.
   - Output: `4`

3. **Example 3**:
   - Input: `s = "luffy is still joyboy"`
   - The last word is `"joyboy"` with a length of `6`.
   - Output: `6`

### Complexity Analysis:

- **Time Complexity**: \(O(n)\), where `n` is the length of the string `s`. The split operation and trimming both traverse the string once.
- **Space Complexity**: \(O(n)\), since the `split()` method may create a list of words that, in the worst case, could be as long as the input string `s`.

This approach is straightforward and efficient for the given problem constraints.
